```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
#loading the data
url="https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
df=pd.read_csv(url)
df.head()
```

```python
#prepare the data 
X=df[['Hours']].values
y=df[['Scores']].values
```

```python
#Polynomial Feature Transformation(degree =2)
poly=PolynomialFeatures(degree=2)
X_poly=poly.fit_transform(X)
```

```python
#preapare model
model=LinearRegression()
model.fit(X_poly,y)
```

```python
#predict
y_pred=model.predict(X_poly)
```

```python
#evaluate
print("R2 Score:",r2_score(y,y_pred))
print("Mean Absolute Error:",mean_absolute_error(y,y_pred))
```

```python
#visualizing the Regression Line
plt.scatter(X,y,color='blue', label='Actual data')
X_sorted=np.sort(X, axis=0)
y_sorted_pred=model.predict(poly.transform(X_sorted))
plt.plot(X_sorted, y_sorted_pred, color='red', label='Polynomial Regression')#Regression Line
plt.title('Polynomial Regression (Hours vs Scores)')
plt.xlabel("Hours Studied")
plt.ylabel("Scores")                           
plt.show()
```
