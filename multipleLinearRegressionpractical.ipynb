```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(1)
study_hours=np.random.randint(1,10,50)
attendence=np.random.randint(50,100,50)
sleep_hours=np.random.randint(5,9,50)
score=(5* study_hours +0.5 * attendence +2*sleep_hours+np.random.normal(0,10,50))
df=pd.DataFrame({
    'study_hours':study_hours,
    'attendence':attendence,
    'sleep_hours':sleep_hours,
    'score':score
})
print(df)
```

```python

X=df[['study_hours','attendence','sleep_hours']]
Y=df[['score']]
model =LinearRegression()
model.fit(X, Y)
```

```python
print("Coefficients:",model.coef_)
print("Intercept:",model.intercept_)
```

```python
plt.scatter(X,Y,color='orange')
plt.plot(X, model.predict(X), color='blue')
plt.xlabel("study_hours and attendence and sleep_hours")
plt.ylabel("scores")
plt.title("multiple Linear Regression")
plt.show()
```
